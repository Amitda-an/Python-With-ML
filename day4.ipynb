```python
def sniplot(flip=1):

  x=np.linspace(0, 15, 100)

  for i in range (1 , 5):

      plt.plot(x,np.sin(x+i* .5) *(7-i) * flip)
sns.set_style("white")
sns.set_palette("husl")
sniplot()
plt.show()
```

```python
def sniplot(flip=1):

  x=np.linspace(0, 15, 100)

  for i in range (1 , 5):

      plt.plot(x,np.cos(x+i* .5) *(7-i) * flip)
sns.set_style("white")
sns.set_palette("husl")
sniplot()
plt.show()
```
